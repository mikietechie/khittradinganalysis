```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
with open("input.xlsx", encoding='utf-8', errors="ignore") as f:
    df = pd.read_excel(f.buffer, "工作表1")
    df = df.drop(columns="winning percentage")
```

```python
xcolumns = df.columns[2:6]
ycolumns = df.columns[2:]
count = 0
for xcolumn in xcolumns:
    for ycolumn in ycolumns:
        if xcolumn == ycolumn:
            continue
        count += 1
        _df = df.sort_values(xcolumn)
        plt.scatter(x=df[xcolumn], y=df[ycolumn])
        plt.title(f"{xcolumn}  to {ycolumn}")
        plt.show()
```
